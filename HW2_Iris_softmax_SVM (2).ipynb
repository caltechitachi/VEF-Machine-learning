```
#===============================================================================
# Part (a) - Loading Iris dataset
#===============================================================================

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()


# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                              test_size=0.5, random_state=0)

# Check the size of data and labels
print("Training data shape: {}".format(x_train.shape))
print("Training label shape: {}".format(y_train.shape))
print("Testing data shape: {}".format(x_test.shape))
print("Testing label shape: {}".format(y_test.shape))


#Check if sets balanced
print ('Test  labels 0: {}, 1: {}, 2: {}'.format(np.sum(y_train == 0), 
                                  np.sum(y_train == 1), np.sum(y_train == 2) ) )
print ('Train labels 0: {}, 1: {}, 2: {}'.format(np.sum(y_test == 0), 
                                  np.sum(y_test == 1), np.sum(y_test == 2) ) )
```

```
#===============================================================================
# Part (b) Softmax regression
#===============================================================================

#!pip install pandas_ml  # Uncomment this if you haven't installed pandas_ml

from sklearn.linear_model import LogisticRegression # For softmax regression
from sklearn.metrics import accuracy_score
from pandas_ml import ConfusionMatrix

# train
softmax_reg = LogisticRegression(C=100, solver = 'lbfgs', multi_class = 'multinomial')
softmax_reg.fit(x_train, y_train)

# test
y_pred = softmax_reg.predict(x_test)

# Show confusion matrix
confusion_matrix = ConfusionMatrix(y_test, y_pred)
print("Confusion matrix:\n%s" % confusion_matrix)
print ("Accuracy: %.2f %%" %(100*accuracy_score(y_test, y_pred)))

confusion_matrix.plot()
```

```
#===============================================================================
# Part (c) Support Vector Machines with linear kernel
#===============================================================================

from sklearn.svm import SVC

# train

svm_poly = SVC(kernel = 'linear', C=1) 
svm_poly.fit(x_train, y_train)

# test
y_pred = svm_poly.predict(x_test)

# Show confusion matrix
confusion_matrix = ConfusionMatrix(y_test, y_pred)
print("Confusion matrix:\n%s" % confusion_matrix)
print ("Accuracy: %.2f %%" %(100*accuracy_score(y_test, y_pred)))

confusion_matrix.plot()
```

```
#===============================================================================
# Part (d) Support Vector Machines with polynomial kernel
#===============================================================================

# train

svm_poly = SVC(kernel = 'poly', degree=2, gamma=1, coef0=0)
svm_poly.fit(x_train, y_train)

# test
y_pred = svm_poly.predict(x_test)

# Show confusion matrix
confusion_matrix = ConfusionMatrix(y_test, y_pred)
print("Confusion matrix:\n%s" % confusion_matrix)
print ("Accuracy: %.2f %%" %(100*accuracy_score(y_test, y_pred)))

confusion_matrix.plot()
```

```
#===============================================================================
# Support Vector Machines with RBF kernel
#===============================================================================

# train 
svm_rbf = SVC(C=5, gamma=0.1)
svm_rbf.fit(x_train, y_train)

# test
y_pred = svm_rbf.predict(x_test)

# Show confusion matrix
confusion_matrix = ConfusionMatrix(y_test, y_pred)
print("Confusion matrix:\n%s" % confusion_matrix)
print ("Accuracy: %.2f %%" %(100*accuracy_score(y_test, y_pred)))

confusion_matrix.plot()
```
